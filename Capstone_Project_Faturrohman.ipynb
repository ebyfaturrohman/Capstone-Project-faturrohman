```python
import pandas as pd

df = pd.read_csv("Shopee_Sampled_Reviews.csv")
df.columns = df.columns.str.strip()  # hapus spasi tersembunyi
df = df[['reviewId', 'userName', 'content', 'score']]
df = df.dropna(subset=['content', 'score'])  # buang yang kosong
df.head()
```

```python
def label_sentimen(score):
    if score >= 4:
        return 'pujian'
    elif score == 3:
        return 'netral'
    else:
        return 'keluhan'

df['label'] = df['score'].apply(label_sentimen)
df[['content', 'score', 'label']].sample(5)
```

```python
import re

def bersihkan(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)  # hapus simbol
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['clean_review'] = df['content'].apply(bersihkan)
df[['clean_review', 'label']].head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='label')
plt.title("Distribusi Kategori Komentar Pelanggan Shopee")
plt.xlabel("Kategori Komentar")
plt.ylabel("Jumlah")
plt.show()
```

```python
!pip install -q transformers
from transformers import pipeline

summarizer = pipeline("summarization", model="csebuetnlp/mT5_multilingual_XLSum")

keluhan_text = " ".join(df[df['label'] == 'keluhan']['clean_review'].sample(30).tolist())
keluhan_text = keluhan_text[:1000]  # batas input

ringkasan = summarizer(keluhan_text, max_length=100, min_length=30, do_sample=False)
print("🧾 Ringkasan Keluhan Pelanggan:\n")
print(ringkasan[0]['summary_text'])
```
